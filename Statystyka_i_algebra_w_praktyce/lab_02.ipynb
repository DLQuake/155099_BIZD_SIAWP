```python
# Import bibliotek
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import describe
from scipy.stats import binom
from scipy.stats import norm
```

### 1. Dla zmiennych losowych przedstawionych w tabeli obliczyć podstawowe statystyki

|       Wartość      |          1         |          2         |          3         |          4         |          5         |          6         |
|--------------------|--------------------|--------------------|--------------------|--------------------|--------------------|--------------------|
| Prawdopodobieństwo |         1/6        |         1/6        |         1/6        |         1/6        |         1/6        |         1/6        |

```python
wartosci = [1, 2, 3, 4, 5, 6]
prawdopodobienstwo = [1/6, 1/6, 1/6, 1/6, 1/6, 1/6]

srednia = sum(x * p for x, p in zip(wartosci, prawdopodobienstwo))
wariancja = sum(p * (x - srednia) ** 2 for x, p in zip(wartosci, prawdopodobienstwo))
odchylenie_standardowe = wariancja ** 0.5

print(f"Srednia: {srednia}")
print(f"Wariancja: {wariancja}")
print(f"Odchylenie standardowe: {odchylenie_standardowe}")
```

### 2. Wygeneruj próby dla n=100 dla następujących rozkładów: Bernoulliego, Dwumianowego, Poissona

```python
# Rozkład Bernoulliego
p_bernoulli = 0.5  # prawdopodobieństwo sukcesu (np. rzut monetą)
prob_bernoulli = np.random.choice([0, 1], size=100, p=[1-p_bernoulli, p_bernoulli])

# Rozkład Dwumianowy
n_dwumianowy = 10  # liczba prób
p_dwumianowy = 0.3  # prawdopodobieństwo sukcesu
prob_dwumianowy = np.random.binomial(n_dwumianowy, p_dwumianowy, size=100)

# Rozkład Poissona
lambda_poisson = 3  # intensywność
prob_poisson = np.random.poisson(lambda_poisson, size=100)

# Wyniki
print("Próby z rozkładu Bernoulliego:")
print(prob_bernoulli)

print("\nPróby z rozkładu Dwumianowego:")
print(prob_dwumianowy)

print("\nPróby z rozkładu Poissona:")
print(prob_poisson)
```

### 3. Policz dla nich statystyki podstawowe (średnią, wariancję, kurtozę i skośność).

```python
# Rozkład Bernoulliego
stats_bernoulli = describe(prob_bernoulli)
print("\nStatystyki dla rozkładu Bernoulliego:")
print(stats_bernoulli)

# Rozkład Dwumianowy
stats_dwumianowy = describe(prob_dwumianowy)
print("\nStatystyki dla rozkładu Dwumianowego:")
print(stats_dwumianowy)

# Rozkład Poissona
stats_poisson = describe(prob_poisson)
print("\nStatystyki dla rozkładu Poissona:")
print(stats_poisson)
```

### 4. Dla rozkładów Bernoulliego, Dwumianowego i Poissona narysuj wykres składający się z 3 części, gdzie każda część będzie zawierać wykres rozkładu prawdopodobieństwa

```python
# Wykresy
fig, axs = plt.subplots(3, 1, figsize=(10, 12))

# Wykres dla rozkładu Bernoulliego
axs[0].hist(prob_bernoulli, bins=[-0.5, 0.5, 1.5], edgecolor='black', align='mid', rwidth=0.8)
axs[0].set_title('Rozkład Bernoulliego')
axs[0].set_xticks([0, 1])
axs[0].set_xlabel('Wartość')
axs[0].set_ylabel('Częstość')

# Wykres dla rozkładu Dwumianowego
axs[1].hist(prob_dwumianowy, bins=np.arange(-0.5, n_dwumianowy + 1.5, 1), edgecolor='black', align='mid', rwidth=0.8)
axs[1].set_title('Rozkład Dwumianowy')
axs[1].set_xticks(np.arange(0, n_dwumianowy + 1))
axs[1].set_xlabel('Liczba sukcesów')
axs[1].set_ylabel('Częstość')

# Wykres dla rozkładu Poissona
axs[2].hist(prob_poisson, bins=np.arange(-0.5, max(prob_poisson) + 1.5, 1), edgecolor='black', align='mid', rwidth=0.8)
axs[2].set_title('Rozkład Poissona')
axs[2].set_xticks(np.arange(0, max(prob_poisson) + 1))
axs[2].set_xlabel('Liczba zdarzeń')
axs[2].set_ylabel('Częstość')

plt.tight_layout()
plt.show()
```

### 5. Dla rozkładu Dwumianowego wygeneruj rozkład prawdopodobieństwa dla n = 20, k = 0, …, 20 oraz p = 0.4. Sprawdź, czy suma prawdopodobieństw wygenerowana dla rozkładu dwumianowego jest równa 1.

```python
# Parametry rozkładu dwumianowego
n_dwumianowy = 20  # liczba prób
p_dwumianowy = 0.4  # prawdopodobieństwo sukcesu

# Generowanie rozkładu prawdopodobieństwa
rozklad_dwumianowy = binom.pmf(np.arange(0, n_dwumianowy+1), n_dwumianowy, p_dwumianowy)

# Sprawdzenie, czy suma prawdopodobieństw wynosi 1
suma_prawdopodobienstw = np.sum(rozklad_dwumianowy)

# Wyniki
print("Rozkład prawdopodobieństwa dla rozkładu Dwumianowego:")
print(rozklad_dwumianowy)

print("\nSuma prawdopodobieństw:", suma_prawdopodobienstw)
```

### 6. Wygeneruj 100 danych dla rozkładu normalnego o średniej równej 0 i odchyleniu standardowym równym 2. Wyznacz wszystkie statystyki podstawowe – czy są one równe z wartościami teoretycznymi? Sprawdź, czy zwiększenie liczby danych zwiększy dokładność wyliczeń statystyk opisowych.

```python
# Parametry rozkładu normalnego
srednia = 0
odchylenie_standardowe = 2

# Generowanie 100 danych dla rozkładu normalnego
dane_normalne = np.random.normal(srednia, odchylenie_standardowe, size=100)

# Obliczenia statystyk opisowych dla 100 danych
statystyki_100_danych = describe(dane_normalne)
print("Statystyki opisowe dla 100 danych:")
print(statystyki_100_danych)

# Zwiększenie liczby danych do 1000
dane_normalne_1000 = np.random.normal(srednia, odchylenie_standardowe, size=1000)

# Obliczenia statystyk opisowych dla 1000 danych
statystyki_1000_danych = describe(dane_normalne_1000)
print("\nStatystyki opisowe dla 1000 danych:")
print(statystyki_1000_danych)
```

### 7. Narysować na jednym wykresie histogram dla rozkładu normalnego o parametrach: średnia = 1, odchylenie =2, wykres dla rozkładu standardowego, oraz wykres gęstości dla średniej równej -1 oraz odchylenia równego 0.5

```python
# Parametry rozkładu normalnego dla histogramu
srednia_hist = 1
odchylenie_standardowe_hist = 2

# Parametry rozkładu normalnego dla wykresu gęstości
srednia_gestosc = -1
odchylenie_standardowe_gestosc = 0.5

# Generowanie danych dla histogramu
dane_hist = np.random.normal(srednia_hist, odchylenie_standardowe_hist, size=1000)

# Wykres histogramu
plt.hist(dane_hist, bins=30, density=True, alpha=0.7, color='blue', label='Histogram (μ=1, σ=2)')

# Generowanie danych dla wykresu gęstości
x_gestosc = np.linspace(-5, 3, 1000)
gestosc = norm.pdf(x_gestosc, srednia_gestosc, odchylenie_standardowe_gestosc)

# Wykres gęstości
plt.plot(x_gestosc, gestosc, 'r-', label='Gęstość (μ=-1, σ=0.5)')

# Ustawienia wykresu
plt.title('Histogram i Wykres Gęstości')
plt.xlabel('Wartość')
plt.ylabel('Prawdopodobieństwo')
plt.legend()

# Wyświetlenie wykresu
plt.show()
```
